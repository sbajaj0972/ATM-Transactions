```python
import tensorflow as tf
from tensorflow import keras
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc_x_train = sc.fit_transform(x_train)
sc_x_test = sc.fit_transform(x_test)
```

```python
x_train1 = pd.DataFrame(sc_x_train)
x_test1 = pd.DataFrame(sc_x_test)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(sc_x_train, y_train, test_size=0.25, random_state=101)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Build DNN Architecture


# Setup Sequential API for building Deep Neural Network 
dnn = tf.keras.models.Sequential()
dnn.add(tf.keras.layers.Dense(units=15, activation='relu'))

dnn.add(tf.keras.layers.Dense(units=10, activation='relu'))

dnn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

```python
# Connect with training dataset by using multi layer perceptron

dnn.compile(optimizer = 'adam', loss='binary_crossentropy', metrics=['accuracy'])
dnn.fit(x_train, y_train, batch_size = 100, epochs=20)
```

```python
dnn.fit(x_test, y_test, batch_size=100, epochs=10)
```
